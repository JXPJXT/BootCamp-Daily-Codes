```python
class University:
    def __init__(self, name):
        self.name = name
        self.departments = []

    def add_department(self, department):
        self.departments.append(department)
    
    def get_departments(self):
        return self.departments

    def __str__(self):
        return f"University: {self.name}"


class Department:
    def __init__(self, name):
        self.name = name
        self.professors = []
        self.students = []

    def add_professor(self, professor):
        self.professors.append(professor)

    def add_student(self, student):
        self.students.append(student)

    def get_professors(self):
        return self.professors
    
    def get_students(self):
        return self.students

    def __str__(self):
        return f"Department: {self.name}"


class Professor:
    def __init__(self, name, department):
        self.name = name
        self.department = department
        department.add_professor(self)

    def __str__(self):
        return f"Professor: {self.name}, Department: {self.department.name}"


class Student:
    def __init__(self, name, department):
        self.name = name
        self.department = department
        department.add_student(self)

    def __str__(self):
        return f"Student: {self.name}, Department: {self.department.name}"


# Example usage:
if __name__ == "__main__":
    # Create a university
    uni = University("Lovely Professional University")

    # Create departments
    cs_dept = Department("Computer Science")
    ee_dept = Department("Electrical Engineering")

    # Add departments to the university
    uni.add_department(cs_dept)
    uni.add_department(ee_dept)

    # Create professors
    prof_rahul = Professor("Rahul Sharma", cs_dept)
    prof_anita = Professor("Anita Desai", ee_dept)

    # Create students
    student_ravi = Student("Ravi Kumar", cs_dept)
    student_sneha = Student("Sneha Singh", ee_dept)

    # Print university details
    print(uni)
    for dept in uni.get_departments():
        print(dept)
        for prof in dept.get_professors():
            print(prof)
        for student in dept.get_students():
            print(student)
```

```python
class University:
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.departments = []

    def add_department(self, department):
        self.departments.append(department)

    def __str__(self):
        dept_names = ', '.join([dept.name for dept in self.departments])
        return f"University: {self.name}\nLocation: {self.location}\nDepartments: {dept_names}"


class Department:
    def __init__(self, name, building):
        self.name = name
        self.building = building
        self.professors = []
        self.students = []

    def add_professor(self, professor):
        self.professors.append(professor)

    def add_student(self, student):
        self.students.append(student)

    def __str__(self):
        prof_names = ', '.join([prof.name for prof in self.professors])
        student_names = ', '.join([student.name for student in self.students])
        return (f"Department: {self.name}\nBuilding: {self.building}\n"
                f"Professors: {prof_names}\nStudents: {student_names}")


class Professor:
    def __init__(self, name, designation):
        self.name = name
        self.designation = designation

    def __str__(self):
        return f"Professor: {self.name}, Designation: {self.designation}"


class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number

    def __str__(self):
        return f"Student: {self.name}, Roll Number: {self.roll_number}"


if __name__ == "__main__":
    uni = University("Lovely Professional University", "Phagwara, Punjab")

    cs_dept = Department("Computer Science", "Block 30")
    ee_dept = Department("Electrical Engineering", "Block 25")

    uni.add_department(cs_dept)
    uni.add_department(ee_dept)

    prof_rahul = Professor("Rahul Sharma", "Associate Professor")
    prof_anita = Professor("Anita Desai", "Professor")

    cs_dept.add_professor(prof_rahul)
    ee_dept.add_professor(prof_anita)

    student_ravi = Student("Ravi Kumar", "CS001")
    student_sneha = Student("Sneha Singh", "EE001")

    cs_dept.add_student(student_ravi)
    ee_dept.add_student(student_sneha)

    print(uni)
    print("\nDepartments Details:")
    for dept in uni.departments:
        print(dept)
```

```python
class University:
    def __init__(self, name, location):
        self.name = name
        self.location = location
        self.departments = []
        print(f"University {self.name} in {self.location} has been created.")

    def add_department(self, department):
        self.departments.append(department)

    def __str__(self):
        dept_names = ', '.join([dept.name for dept in self.departments])
        return f"University: {self.name}\nLocation: {self.location}\nDepartments: {dept_names}"

    def __del__(self):
        print(f"University {self.name} in {self.location} has been deleted.")


class Department:
    def __init__(self, name, building):
        self.name = name
        self.building = building
        self.professors = []
        self.students = []
        print(f"Department {self.name} in {self.building} has been created.")

    def add_professor(self, professor):
        self.professors.append(professor)

    def add_student(self, student):
        self.students.append(student)

    def __str__(self):
        prof_names = ', '.join([prof.name for prof in self.professors])
        student_names = ', '.join([student.name for student in self.students])
        return (f"Department: {self.name}\nBuilding: {self.building}\n"
                f"Professors: {prof_names}\nStudents: {student_names}")

    def __del__(self):
        print(f"Department {self.name} in {self.building} has been deleted.")


class Professor:
    def __init__(self, name, designation):
        self.name = name
        self.designation = designation
        print(f"Professor {self.name}, Designation: {self.designation}, has been created.")

    def __str__(self):
        return f"Professor: {self.name}, Designation: {self.designation}"

    def __del__(self):
        print(f"Professor {self.name}, Designation: {self.designation}, has been deleted.")


class Student:
    def __init__(self, name, roll_number):
        self.name = name
        self.roll_number = roll_number
        print(f"Student {self.name}, Roll Number: {self.roll_number}, has been created.")

    def __str__(self):
        return f"Student: {self.name}, Roll Number: {self.roll_number}"

    def __del__(self):
        print(f"Student {self.name}, Roll Number: {self.roll_number}, has been deleted.")


if __name__ == "__main__":
    uni = University("Lovely Professional University", "Phagwara, Punjab")

    cs_dept = Department("Computer Science", "Block 30")
    ee_dept = Department("Electrical Engineering", "Block 25")

    uni.add_department(cs_dept)
    uni.add_department(ee_dept)

    prof_rahul = Professor("Rahul Sharma", "Associate Professor")
    prof_anita = Professor("Anita Desai", "Professor")

    cs_dept.add_professor(prof_rahul)
    ee_dept.add_professor(prof_anita)

    student_ravi = Student("Ravi Kumar", "CS001")
    student_sneha = Student("Sneha Singh", "EE001")

    cs_dept.add_student(student_ravi)
    ee_dept.add_student(student_sneha)

    print(uni)
    print("\nDepartments Details:")
    for dept in uni.departments:
        print(dept)
```
