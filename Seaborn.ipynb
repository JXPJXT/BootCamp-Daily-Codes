# Seaborn

## 1. Introduction to Seaborn

Seaborn is a Python visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive statistical graphics. Seaborn integrates seamlessly with Pandas dataframes, simplifying the process of plotting data.

## 3. Basic Plots
### 3.1. Scatter Plot
Scatter plots are useful for displaying the relationship between two variables.

````python
import seaborn as sns
import matplotlib.pyplot as plt

# Load example dataset
tips = sns.load_dataset("tips")

# Create scatter plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
````

### 3.2. Line Plot
Line plots are used to visualize data points connected by a line.



````python
# Create line plot
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
````

### 3.3. Bar Plot
Bar plots are useful for showing the distribution of categorical data.

````python
# Create bar plot
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
````

### 3.4. Histogram
Histograms are used to represent the distribution of a variable.

````python
# Create histogram
sns.histplot(data=tips, x="total_bill")
plt.show()
````

### 3.5. Box Plot
Box plots display the distribution of data based on a five-number summary.

````python
# Create box plot
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
````

## 4. Customization
Customizing plots can improve their readability and visual appeal.

````python
# Customizing a bar plot
sns.set(style="whitegrid")
sns.barplot(data=tips, x="day", y="total_bill", palette="Blues_d")
plt.show()
````

### Q1. Create a scatter plot showing the relationship between total bill and tip amount, colored by day.

````python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="day")
plt.show()
````

### Q2. Use a pair plot to visualize the relationships between different variables in the tips dataset, with different colors for different times of day.

````python
sns.pairplot(tips, hue="time")
plt.show()

````
