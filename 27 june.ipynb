```python
import pandas as pd

data = {
    'Product': ['Laptop', 'Tablet', 'Smartphone', 'Desktop'],
    'Price': [1200, 300, 800, 600],
    'Quantity': [5, 15, 25, 10]
}

df = pd.DataFrame(data)

df.head()
df.info()
df.describe()

df['Price']
df[['Product', 'Quantity']]

df.loc[1]
df.iloc[3]
df[df['Price'] > 500]

df['Discounted_Price'] = df['Price'] * 0.9
df['Stock_Value'] = df['Price'] * df['Quantity']
df.drop(columns=['Discounted_Price'], inplace=True)
df.sort_values(by='Quantity')

df.to_csv('products.csv', index=False)
df_read = pd.read_csv('products.csv')
pd.concat([df, df_read])
```

```python
import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 24, 35, 32],
    'City': ['New York', 'Paris', 'Berlin', 'London']
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# Exploring DataFrame attributes
print("Columns:", df.columns)
print("Index:", df.index)
print("Data types:")
print(df.dtypes)
print("Shape:", df.shape)

# Viewing DataFrame content
print("First 3 rows:")
print(df.head(3))
print("Last 2 rows:")
print(df.tail(2))

# Selecting columns
print("Age column:")
print(df['Age'])
print("Name and City columns:")
print(df[['Name', 'City']])

# Selecting rows
print("First row (using loc):")
print(df.loc[0])
print("Third row (using iloc):")
print(df.iloc[2])

# Filtering rows
print("Rows where Age > 30:")
print(df[df['Age'] > 30])

# Adding a new column
df['Country'] = ['USA', 'France', 'Germany', 'UK']
print("DataFrame with new column:")
print(df)

# Removing a column
df.drop(columns=['Country'], inplace=True)
print("DataFrame after removing the Country column:")
print(df)

# Sorting the DataFrame
df_sorted = df.sort_values(by='Age')
print("DataFrame sorted by Age:")
print(df_sorted)

# Basic statistics
print("Statistics summary:")
print(df.describe())

# Exporting the DataFrame to a CSV file
df.to_csv('data.csv', index=False)
print("DataFrame exported to data.csv")

# Reading the DataFrame from a CSV file
df_read = pd.read_csv('data.csv')
print("DataFrame read from data.csv:")
print(df_read)

# Concatenating DataFrames
df_concat = pd.concat([df, df_read])
print("Concatenated DataFrame:")
print(df_concat)
```

```python
import pandas as pd

# Creating a DataFrame from a dictionary of lists
data = {
    'Product': ['Laptop', 'Tablet', 'Smartphone', 'Desktop'],
    'Price': [1200, 300, 800, 600],
    'Quantity': [5, 15, 25, 10]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)

# DataFrame attributes
print("Columns:", df.columns)
print("Index:", df.index)
print("Data types:")
print(df.dtypes)
print("Shape:", df.shape)

# Viewing DataFrame content
print("First 2 rows:")
print(df.head(2))
print("Last 3 rows:")
print(df.tail(3))

# Selecting columns
print("Price column:")
print(df['Price'])
print("Product and Quantity columns:")
print(df[['Product', 'Quantity']])

# Selecting rows
print("Second row (using loc):")
print(df.loc[1])
print("Fourth row (using iloc):")
print(df.iloc[3])

# Filtering rows
print("Rows where Price > 500:")
print(df[df['Price'] > 500])

# Adding a new column
df['Discounted_Price'] = df['Price'] * 0.9
print("DataFrame with Discounted_Price column:")
print(df)

# Removing a column
df.drop(columns=['Discounted_Price'], inplace=True)
print("DataFrame after removing the Discounted_Price column:")
print(df)

# Sorting the DataFrame
df_sorted = df.sort_values(by='Quantity')
print("DataFrame sorted by Quantity:")
print(df_sorted)

# Basic statistics
print("Statistics summary:")
print(df.describe())

# Exporting the DataFrame to a CSV file
df.to_csv('products.csv', index=False)
print("DataFrame exported to products.csv")

# Reading the DataFrame from a CSV file
df_read = pd.read_csv('products.csv')
print("DataFrame read from products.csv:")
print(df_read)

# Concatenating DataFrames
df_concat = pd.concat([df, df_read])
print("Concatenated DataFrame:")
print(df_concat)

# Grouping and Aggregation
grouped = df.groupby('Product').sum()
print("Grouped by Product and summed:")
print(grouped)

# Pivot Tables
pivot_table = df.pivot_table(values='Price', index='Product', aggfunc='mean')
print("Pivot table with mean Price per Product:")
print(pivot_table)

# Merging DataFrames
data_1 = {'Name': ['Alice', 'Bob', 'Charlie'], 'Age': [25, 30, 35]}
data_2 = {'Name': ['Alice', 'Bob', 'Charlie'], 'City': ['New York', 'Los Angeles', 'Chicago']}
df1 = pd.DataFrame(data_1)
df2 = pd.DataFrame(data_2)
df_merged = pd.merge(df1, df2, on='Name')
print("Merged DataFrame:")
print(df_merged)

# Handling Missing Data
data_with_nan = {'A': [1, 2, None], 'B': [None, 2, 3]}
df_nan = pd.DataFrame(data_with_nan)
print("DataFrame with NaN values:")
print(df_nan)
print("DataFrame after dropping rows with NaN values:")
print(df_nan.dropna())
print("DataFrame after filling NaN values with 0:")
print(df_nan.fillna(0))

# Applying Functions
df['Price_with_tax'] = df['Price'].apply(lambda x: x * 1.1)
print("DataFrame with Price_with_tax column:")
print(df)

# Iterating over rows
for index, row in df.iterrows():
    print(f"Index: {index}, Product: {row['Product']}, Price: {row['Price']}, Quantity: {row['Quantity']}")

# Renaming Columns
df_renamed = df.rename(columns={'Product': 'Item', 'Price': 'Cost'})
print("DataFrame with renamed columns:")
print(df_renamed)
```
